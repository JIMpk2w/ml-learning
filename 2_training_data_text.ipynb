```python
import imblearn

ros = imblearn.over_sampling.RandomOverSampler(sampling_strategy="not majority", random_state=69420)
train_X, train_y = ros.fit_resample(train_data, train_data["category"])
train_X["category"].value_counts()
```

```python
train_data["organization"].value_counts().head(10)
```

```python
rus = imblearn.under_sampling.RandomUnderSampler(sampling_strategy="not minority", random_state=69420)
train_X, train_y = rus.fit_resample(train_data, train_data["organization"])
train_X["organization"].value_counts().head(10)
```
